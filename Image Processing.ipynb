```python
# Resizing the image using pillow library
from PIL import Image
```

```python
img = Image.open('dog.jpg')
```

```python
img_resized = img.resize((200,200))
```

```python
img_resized.save('dogimageresized.jpg')
```

```python
img1 = mpimg.imread('dogimageresized.jpg')
```

```python
img1.shape
```

```python
# Convert RGB images to greyscale using OPENCV
import cv2
```

```python
img = cv2.imread('dog.jpg')
```

```python
img.shape
```

```python
grayscale_image = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
```

```python
grayscale_image.shape
```

```python
imgplot = plt.imshow(grayscale_image)
```

```python
# saving grayscale image
cv2.imwrite('doggreyscaleimg.jpg',grayscale_image)
```
